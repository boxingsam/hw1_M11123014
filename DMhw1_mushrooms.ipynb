```python
#將分類的欄位另外提出來
X = df.drop('class', axis=1)
y = df['class']
X.head()
```

```python
#get_dummies會依據其變數名稱自動新增新行名。
dum_X = pd.get_dummies(X)
dum_X = pd.DataFrame(dum_X)
dum_X
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
label_y = labelencoder.fit_transform(y)
label_y
```

```python
# from sklearn.preprocessing import OneHotEncoder
# onehotencoder = OneHotEncoder()
# onehot_X = onehotencoder.fit_transform(X).toarray()
# pd.DataFrame(onehot_X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dum_X, label_y, test_size=0.3, random_state=42)
```

```python
X_test.shape
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
train_list=[]
test_list=[]
for i in range(1,16):
    model= DecisionTreeClassifier(criterion='gini',max_depth=i,random_state=42)  
    model.fit(X_train, y_train)
    score_test=model.score(X_test, y_test)
    score_train=model.score(X_train, y_train)
    test_list.append(score_test)
    train_list.append(score_train)
result={
    "train":train_list,
    "test":test_list
}
df_result=pd.DataFrame(result)
```

```python
import matplotlib
matplotlib.use('TKAgg') 
%matplotlib inline
import matplotlib.pyplot as plt


plt.plot(train_list,color=(255/255,100/255,100/255))
plt.plot(test_list,color=(100/255,100/255,255/255))

plt.ylabel("Accuracy")
plt.xlabel("Depth_of_tree")
plt.title("Adult_dataset_accuracy")
plt.xlim([0,16])
plt.grid(True)

plt.show()
```

```python
print(df_result,'\n',df_result.max())
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_gini=DecisionTreeClassifier(criterion='gini',max_depth=7,random_state=42)
dt_gini.fit(X_train,y_train)
```

```python
#訓練集正確率
dt_gini.score(X_train,y_train)
```

```python
#測試集正確率
dt_gini.score(X_test,y_test)
```

```python
y_pred_dt_gini_train=dt_gini.predict(X_train)
y_pred_dt_gini_test=dt_gini.predict(X_test)
print("y_pred_dt_gini_train:",y_pred_dt_gini_train)
print("y_pred_dt_gini_test:",y_pred_dt_gini_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("訓練資料的混淆矩陣:")
print(confusion_matrix(y_train,y_pred_dt_gini_train))
print("測試資料的混淆矩陣:")
print(confusion_matrix(y_test,y_pred_dt_gini_test))
```

```python
print("訓練資料的準確率:",accuracy_score(y_train,y_pred_dt_gini_train))
print("測試資料的準確率:",accuracy_score(y_test,y_pred_dt_gini_test))
```
